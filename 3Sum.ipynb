# 3Sum
Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

Notice that the solution set must not contain duplicate triplets.

Example 1:

    Input: nums = [-1,0,1,2,-1,-4]
    Output: [[-1,-1,2],[-1,0,1]]

Explanation: 

    nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0.
    nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0.
    nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0.
    The distinct triplets are [-1,0,1] and [-1,-1,2].
    Notice that the order of the output and the order of the triplets does not matter.

Example 2:

    Input: nums = [0,1,1]
    Output: []

Explanation: The only possible triplet does not sum up to 0.
Example 3:

    Input: nums = [0,0,0]
    Output: [[0,0,0]]

Explanation: The only possible triplet sums up to 0.
 

Constraints:

    3 <= nums.length <= 3000
    -105 <= nums[i] <= 105

```python
from typing import List

class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        res = set()

        # Phân chia mảng thành các số âm, số dương và số không
        n, p, z = [], [], []
        for num in nums:
            if num > 0:
                p.append(num)
            elif num < 0:
                n.append(num)
            else:
                z.append(num)

        # Tạo các tập hợp cho số âm và số dương để tra cứu O(1)
        N, P = set(n), set(p)

        # Kiểm tra các trường hợp có số không
        if z:
            for num in P:
                if -num in N:
                    res.add((-num, 0, num))

        # Nếu có ít nhất ba số không, thêm bộ ba (0, 0, 0) vào kết quả
        if len(z) >= 3:
            res.add((0, 0, 0))

        # Kiểm tra các cặp số âm và tìm phần bù trong tập hợp số dương
        for i in range(len(n)):
            for j in range(i + 1, len(n)):
                target = -1 * (n[i] + n[j])
                if target in P:
                    res.add(tuple(sorted([n[i], n[j], target])))

        # Kiểm tra các cặp số dương và tìm phần bù trong tập hợp số âm
        for i in range(len(p)):
            for j in range(i + 1, len(p)):
                target = -1 * (p[i] + p[j])
                if target in N:
                    res.add(tuple(sorted([p[i], p[j], target])))

        return [list(triplet) for triplet in res]

sol = Solution()
print(sol.threeSum([-1, 0, 1, 2, -1, -4]))
```



### Giải thích mã

1. **Phân chia mảng thành các số âm, số dương và số không**:
   - Tạo ba danh sách `n`, `p` và `z` để chứa các số âm, số dương và số không từ mảng đầu vào.

2. **Tạo các tập hợp cho số âm và số dương để tra cứu O(1)**:
   - Tạo các tập hợp `N` và `P` từ các danh sách `n` và `p` để có thể tra cứu nhanh chóng.

3. **Kiểm tra các trường hợp có số không**:
   - Nếu có ít nhất một số không, kiểm tra xem có các cặp số âm và số dương sao cho tổng của chúng bằng không.
   - Nếu có ít nhất ba số không, thêm bộ ba `(0, 0, 0)` vào kết quả.

4. **Kiểm tra các cặp số âm và tìm phần bù trong tập hợp số dương**:
   - Với mỗi cặp số âm, tính toán phần bù cần thiết để tổng của bộ ba bằng không và kiểm tra xem phần bù có trong tập hợp số dương hay không.

5. **Kiểm tra các cặp số dương và tìm phần bù trong tập hợp số âm**:
   - Với mỗi cặp số dương, tính toán phần bù cần thiết để tổng của bộ ba bằng không và kiểm tra xem phần bù có trong tập hợp số âm hay không.
