```python
from collections import defaultdict

class Solution:
    def isValidSudoku(self, board):
        # Create defaultdicts to store the values in each row, column, and 3x3 square
        cols = defaultdict(set)
        rows = defaultdict(set)
        square = defaultdict(set) 
        
        # Iterate through each cell in the board
        for r in range(9):
            for c in range(9):
                # Skip empty cells
                if board[r][c] == '.':
                    continue
                
                # Check if the value already exists in the same row, column, or square
                if (
                    board[r][c] in rows[r]
                    or board[r][c] in cols[c]
                    or board[r][c] in square[(r//3, c//3)]
                ):
                    return False
                
                # Add the value to the corresponding row, column, and square sets
                cols[c].add(board[r][c])
                rows[r].add(board[r][c])
                square[(r//3, c//3)].add(board[r][c])
        
        # If all cells pass the validation, return True
        return True
    
    
    
```



### Giải thích 

1. **Khởi tạo defaultdict**:
   - `cols`, `rows`, và `square` đều là các `defaultdict` của các `set`. Điều này giúp chúng ta dễ dàng kiểm tra sự tồn tại của các giá trị mà không cần phải kiểm tra xem khóa đã tồn tại trong từ điển hay chưa.

2. **Duyệt qua bảng Sudoku**:
   - Mã duyệt qua từng ô của bảng 9x9.
   - Nếu ô hiện tại chứa ký tự '.', nó sẽ bỏ qua ô đó.
   - Sau đó, kiểm tra xem giá trị hiện tại đã tồn tại trong hàng, cột, hoặc ô vuông 3x3 tương ứng chưa. Nếu có, trả về `False` vì điều này vi phạm quy tắc Sudoku.

3. **Thêm giá trị vào tập hợp**:
   - Nếu giá trị chưa tồn tại trong hàng, cột hoặc ô vuông 3x3, nó sẽ được thêm vào các tập hợp tương ứng.

4. **Trả về `True` nếu bảng hợp lệ**.





```python
from collections import defaultdict

class Solution:
    def isValidSudoku(self, board):
        # defaultdict cho các cột, hàng và các ô vuông 3x3
        cols = defaultdict(set)
        rows = defaultdict(set)
        square = defaultdict(set) 
        
        for r in range(9):
            for c in range(9):
                # Bỏ qua các ô trống
                if board[r][c] == '.':
                    continue
                
                # Kiểm tra sự tồn tại của giá trị trong hàng, cột, hoặc ô vuông 3x3
                if (
                    board[r][c] in rows[r] or
                    board[r][c] in cols[c] or
                    board[r][c] in square[(r // 3, c // 3)]
                ):
                    return False
                
                # Thêm giá trị vào hàng, cột và ô vuông 3x3 tương ứng
                cols[c].add(board[r][c])
                rows[r].add(board[r][c])
                square[(r // 3, c // 3)].add(board[r][c])
        
        # Trả về True nếu bảng hợp lệ
        return True
```
